```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as ani
import itertools
from scipy.linalg import norm
plt.rc('animation', html='jshtml')

np.set_printoptions(precision=3)
```

```python
dt = 0.005 # tidsstep

number_of_particles = 50
#startdata
x0s = np.random.rand(number_of_particles)
y0s = np.random.rand(number_of_particles)
vx0s = np.cos(np.random.rand(number_of_particles)*2*np.pi)
vy0s= np.sin(np.random.rand(number_of_particles)*2*np.pi)

position = np.array([x0s, y0s])
speed = np.array([vx0s, vy0s])



#antal partikler i lister(startdata)
range_of_particles  = range(number_of_particles)

R = 0.05 #radius partikler

#plot
fig, ax = plt.subplots() #vi laver plottet
partikler, = ax.plot([], [], 'ro')

box = ax.plot([0,1,1,0,0], [0,0,1,1,0], 'k')

#klasse
class particles: #for at undgå problemer med lokale variable
    def __init__(self, position1, speed1, r):
        self.position = position1
        self.speed = speed1
                
        self.n_particles = len(position)
        self.radius = r
        
    def velocity_propagation(self):
        for particle in range_of_particles:
            self.position[0][particle] += self.speed[0][particle]*dt
            self.position[1][particle] += self.speed[1][particle]*dt

    def collision_with_walls(self):
        for particle in range_of_particles:

                #højre side af kassen
            if self.position[0][particle] >= 1:
                self.speed[0][particle] *= -1

                #venstre side af kassen 
            if self.position[0][particle] <= 0:
                self.speed[0][particle] *= -1

                #toppen af kassen
            if self.position[1][particle] >= 1:
                self.speed[1][particle] *= -1

                #bunden af kassen
            if self.position[1][particle] <= 0:
                self.speed[1][particle] *= -1
    
        #partikler.set_data(self.x0s, self.y0s)
    
    def particle_particle_collision(self):
        
        for i,j in itertools.combinations(range(0,self.n_particles),r=2):
            

            d = np.linalg.norm(self.position[i]-self.position[j])**2
            
            if d<=4*self.radius**2:
                # detected collision
                
                # handle collision event. 
                self._handle_particle_collision_event(i,j)
    
    def _handle_particle_collision_event(self,index,other_index):
        
        p1_r = self.position[index]
        p1_v = self.speed[index]
        
        p2_r = self.position[other_index]
        p2_v =self.speed[other_index]
               
        self.speed[index] = 0 
#         p2_v-np.dot((p2_v-p1_v),(p2_r-p1_r))/(np.linalg.norm(p2_r-p1_r))**2*(p2_r-p1_r)
        
        self.speed[other_index] =  0
#         p1_v-np.dot((p1_v-p2_v),(p1_r-p2_r))/(np.linalg.norm(p1_r-p2_r))**2*(p1_r-p2_r)
    
par = particles(position, speed, R)

plt.close()
```

```python
#update funktion
def update(i):

    par.velocity_propagation()
    par.collision_with_walls()
    par.particle_particle_collision()
    
    partikler.set_data(par.position[0], par.position[1])
    return partikler,

plt.close()

anim=ani.FuncAnimation(fig, update, frames=200, interval=10, blit=True)
```
