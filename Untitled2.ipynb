```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as ani
plt.rc('animation', html='jshtml')

np.set_printoptions(precision=3)
```

```python
dt = 0.01
x0s = np.array([0.5, 0.4])
y0s = np.array([0.3, 0.8])
vx0s = np.array([1.2, 1.6])
vy0s = np.array([1.3, 0.9])
a = range(2)

fig, ax = plt.subplots() #vi laver plottet
partikler, = ax.plot([], [], 'ro')# partiklens startposistion

class kasse: #for at undgå problemer med lokale variable
    def __init__(self, xs, ys, vys, vxs):
        self.x0s = xs
        self.y0s = ys
        self.vx0s = vxs
        self.vy0s = vys
        
    def partikelbevægelse(self, i):
        for partikel in a:
            self.x0s[partikel] += self.vx0s[partikel]*dt
            self.y0s[partikel] += self.vy0s[partikel]*dt

            partikler.set_data(self.x0s[partikel], self.y0s[partikel])

            #højre side af kassen
            if self.x0s[partikel] >= 1:
                self.vx0s[partikel] *= -1

            #venstre side af kassen 
            if self.x0s[partikel] <= 0:
                self.vx0s[partikel] *= -1

            #toppen af kassen
            if self.y0s[partikel] >= 1:
                self.vy0s[partikel] *= -1

            #bunden af kassen
            if self.y0s[partikel] <= 0:
                self.vy0s[partikel] *= -1
                
        return (partikler,)

par1 = kasse(x0s, y0s, vx0s, vy0s)
```
