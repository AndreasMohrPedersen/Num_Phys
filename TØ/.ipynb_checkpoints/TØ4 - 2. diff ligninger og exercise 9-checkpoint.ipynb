```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python

trange=[0,10]
t = np.linspace(0,10,100)


dydt = lambda t,y: [y[1], -y[0]]

d = solve_ivp(dydt, trange, [1, 0], t_eval=t)
ts = d.t
xs = d.y[0]
vs = d.y[1]

plt.plot(ts, xs)
plt.plot(ts, vs)
```

```python
trange=[1,6]
t = np.linspace(1,6,100)


dydt = lambda t,y: [y[1], 6*y[0]/(t*(t-4))]

d = solve_ivp(dydt, trange, [21/8, -7/8], t_eval=t)
ts = d.t
xs = d.y[0]
vs = d.y[1]
print(xs[-1])

plt.plot(ts, xs)
plt.plot(ts, vs)
```

```python
# Exercise 9
m = 1
M = 10
rx0 = 1
ry0 = 0
vx0 = 0
vy0 = 2
#y vektor
y0 = [rx0, ry0, vx0, vy0]
t_span = [0,0.8]
t_eval = np.linspace(0, 0.8, 100)


def dydt(t,y):
    rx=y[0]
    ry=y[1]
    vx=y[2]
    vy=y[3]
    drxdt=vx
    drydt=vy
    d=np.sqrt(rx**2+ry**2)
    dvxdt = -M/d**3*rx
    dvydt = -M/d**3*ry
    return [drxdt, drydt, dvxdt, dvydt]

res = solve_ivp(dydt, t_span, y0, t_eval = t_eval)

plt.plot(res.y[0], res.y[1])
plt.plot(0, 0, 'yo', ms="30")
plt.plot(res.y[0][-1], res.y[1][-1], 'r.', ms=20)
plt.grid()

```
