# Exercise 1

```python
import numpy as np
from scipy.integrate import solve_ivp
```

```python
#a)
y_first = [23.0505]
ts = [0, 100]

dydt = lambda t,y: np.exp(-y/20)

mysol = solve_ivp(dydt , ts, y_first)
mysol
```

```python
#b)
tinit = 0
tfinal = 1
trange = [tinit, tfinal]
yinit = [5.8568]

dydt = lambda t,y: y**(5/4)
mysol = solve_ivp(dydt, trange, yinit)
mysol
```

```python
#c)
trange = [0, -10]# backwards integration
yinit = [1.49831]

dydt = lambda t,y: -1/3*y
mysol = solve_ivp(dydt, trange, yinit)
mysol
```
