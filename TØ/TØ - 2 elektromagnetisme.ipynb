```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize
plt.rc('font', size=16)
```

```python
#opgave 1
f= lambda x: x**6-25*x**5+195*x**4-347*x**3-1564*x**2+4908*x-3168
```

```python
#a)
x=np.linspace(-3,11,100)
plt.title('f(x)')
plt.grid(True)
plt.plot(x,f, '--')
```

```python
#b) - omkring x=1
xmin=scipy.optimize.fmin(func=f, x0=1)
xmin,f(xmin)
```

```python
#b) - omkring x=4
xmin=scipy.optimize.fmin(func=f, x0=4)
xmin,f(xmin)
```

```python
#b) - omkring x=10
xmin=scipy.optimize.fmin(func=f, x0=10)
xmin,f(xmin)# x- og y-værdi i minimum
```

```python
#exercise 2
f= lambda x: x**3+x**2-10*x+8
```

```python
#a)
x=np.linspace(-10,10,20)
plt.plot(x,f(x))
plt.plot(x, np.zeros(20))#plotter y=0
x1=scipy.optimize.fsolve(f,-4)
x2=scipy.optimize.fsolve(f,1)
x3=scipy.optimize.fsolve(f,2)
(x1,x2,x3)
plt.plot(x1,0,'H')
plt.plot(x2,0,'o')
plt.plot(x3,0,'o')
```

```python
#b)
f= lambda x: x**3-6*x**2-x+30
```

```python
x=np.linspace(-5,10,50)
plt.plot(x,f(x))
plt.plot(x, np.zeros(50))
#for x0=4
xmin=scipy.optimize.fmin(func=f, x0=4)
xmin,f(xmin)# x- og y-værdi i minimum
#maksimum= min for -f omkring x0=0
xmin=scipy.optimize.fmin(func= lambda x: -f(x), x0=0)
xmin,f(xmin)# x- og y-værdi i maksimum
```
