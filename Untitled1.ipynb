```python
import numpy as np
np.set_printoptions(precision=3)

dt = 0.1#tidsinterval?

xs = np.array([0.607, 0.757,  0.638, 0.635, 0.190])#x-koordinater?
vxs = np.array([-0.580, 2.75, -1.044, -1.543,  0.568])#hastighedskomposenter i x-retningen?

print('x:',xs)
print('v:',vxs)
```

```python
xs += vxs * dt
print('x:',xs)
```

```python
print('x%1',xs%1)
```

```python
xstack = np.array(xs)
print('xstack:\n',xstack)
for i in range(2):
    xs += vxs * dt
    xstack = np.vstack((xstack,xs))
    print('xstack:\n',xstack)
```

```python
xs = np.array([0.607, 0.757,  0.638, 0.635, 0.190])
vxs = np.array([-0.580, 2.75, -1.044, -1.543,  0.568])
print('x:',xs)
print('v:',vxs)
```

```python
time_to_hit_wall = (1-xs)/vxs
time_to_hit_wall
```

```python
moving_right = vxs > 0
moving_right
```

```python
np.logical_not(moving_right)
```

```python
LARGE_TIME = 100
assert LARGE_TIME > dt
large_time_if_not_hitting_wall = np.logical_not(moving_right) * LARGE_TIME
large_time_if_not_hitting_wall
```

```python
time_to_hit_wall_or_large_time_if_not = \
np.maximum(time_to_hit_wall,large_time_if_not_hitting_wall)
time_to_hit_wall_or_large_time_if_not
```

```python
which_atom_hits_wall_first = np.argmin(time_to_hit_wall_or_large_time_if_not)
which_atom_hits_wall_first
```

```python
time_to_next_atom_hitting_a_wall = \
    time_to_hit_wall_or_large_time_if_not[which_atom_hits_wall_first]
time_to_next_atom_hitting_a_wall
```

```python
xs += vxs * time_to_next_atom_hitting_a_wall
print('x:',xs)
```

```python
print('v:',vxs)
vxs[which_atom_hits_wall_first] *= -1
print('v:',vxs)
```
