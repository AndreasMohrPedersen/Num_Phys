```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def covid19():
    test = np.random.randint(1, 1000)
    
    if test >= 5:
#         print(0, "Test - Negativ") 
        return False
    elif test < 5:
#         print(1, "Test - Positiv")
        return True


def patientzero(initial_health, initial_victims, day_count):
    
    status = initial_health
    covid19_victims = initial_victims
    
    day_X = []
    covid19_victims_Y = []
    
    current_day = 1
    
    while current_day <= day_count:
        if covid19():
            status += 1
            covid19_victims += 1
            day_X.append(current_day)
            covid19_victims_Y.append(covid19_victims)
            break
            
        else:
            status = status
            covid19_victims += 0
            day_X.append(current_day)
            covid19_victims_Y.append(covid19_victims)
            
        current_day += 1
    
    if status > 0:
        status = "Patient infected"
        
    elif status == 0:
        status = "Patient healthy"
    
#     print("Status after 100 days:", status)
    plt.plot(day_X, covid19_victims_Y)

# result = patientzero(0, 0, 100)

# print(result)

x = 0
while x < 100:
    
    result = patientzero(0, 0, 100)
    
#     current_day = 1
#     number_of_patients = 1
    
#     number_of_patients_Y = []
     
#     day_X = []
    
#     if covid19() > 0:
#         number_of_patients += 1
#         number_of_patients_Y.append(number_of_patients)
#         day_X.append(current_day)
        
#     else:
#         number_of_patients += 0
#         number_of_patients_Y.append(number_of_patients)
#         day_X.append(current_day)
#     current_day +=1
    x += 1
#     plt.plot(day_X, number_of_patients_Y)
    
plt.xlabel("days")
plt.ylabel("number of infected people")
plt.show()
```
